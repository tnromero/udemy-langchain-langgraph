# Setup

```python
from langchain_openai import OpenAI
from langchain_experimental.agents.agent_toolkits import create_python_agent
from langchain_community.tools import GoogleSerperRun
from langchain_community.utilities import GoogleSerperAPIWrapper
from langchain.prompts import PromptTemplate
import yaml
import os
```

```python
with open('config.yaml', 'r') as config_file:
    config = yaml.safe_load(config_file)
os.environ['OPENAI_API_KEY'] = config['OPENAI_API_KEY']
os.environ['SERPER_API_KEY'] = config['SERPER_API_KEY']
```

```python
openai = OpenAI(model_name='gpt-4.1-nano', temperature=0)
```

# Agente de Busca e Resumo

```python
api_wrapper = GoogleSerperAPIWrapper(
    gl="br",
    hl="pt-br",
)

google_search = GoogleSerperRun(api_wrapper=api_wrapper, k=10, gl="br", hl="pt-br", search_type="search")
```

```python
agent_executor = create_python_agent(
    llm=openai,
    tool=google_search,  
    verbose=True
)
```

```python
prompt_template = PromptTemplate(
    input_variables=["query"],
    template="""
    Pesquise na web sobre {query} e forneça um resumo abrangente sobre o assunto.
    """
)
```

```python
query = "Carl Sagan"
prompt = prompt_template.format(query=query)
```

```python
print(prompt)
```

```python
response = agent_executor.invoke(prompt)
```

```python
print("Entrada do agente:", response['input'])
```

```python
print("Saída do agente:", response['output'])
```
